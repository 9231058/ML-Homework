Here we define some useful functions for naive Bayes. Based on these functions, we are calculating the probability of a given test and return the result with a higher probability.

```python
def delta(Y, y):
    return Y == y

def mu(Y, y, X, i, k):
    d = sum(delta(Y[j],y[k]) for j, _ in enumerate(Y))
    n = sum(X[j][i] * delta(Y[j], y[k]) for j, _ in enumerate(Y))
    return n / d

def sigma(Y, y, X, i, k):
    d = sum(delta(Y[j],y[k]) for j, _ in enumerate(Y)) - 1
    n = sum((((X[j][i] - mu(Y, y, X, i, k))) ** 2) * delta(Y[j], y[k]) for j, _ in enumerate(Y))
    return n / d
```
